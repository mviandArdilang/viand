```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

path_ = "/content/drive/MyDrive/Colab Notebooks/Dataset/student_scores.csv"
dataset = pd.read_csv(path_)
```

```
f = plt.figure(figsize=(12,4))

f.add_subplot(1,2,1)
dataset['Hours'].plot(kind='kde')

f.add_subplot(1,2,2)
plt.boxplot(dataset['Hours'])
plt.show()
```

```
f = plt.figure(figsize=(12,4))

f.add_subplot(1,2,1)
dataset['Scores'].plot(kind='kde', c='g')

f.add_subplot(1,2,2)
plt.boxplot(dataset['Scores'])
plt.show()
```

```
plt.scatter(dataset['Hours'], dataset['Scores'])
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Scatter plot Hours vs Scores')
plt.show()
```

```
dataset.corr()
```

```
X = dataset['Hours'].values.reshape(-1,1)
y = dataset['Scores'].values.reshape(-1,1)
```

```
X
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```
lin_reg = LinearRegression()
```

```
lin_reg.fit(X_train, y_train)
```

```
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```
lin_reg.score(X_test, y_test)
```

```
y_prediksi = lin_reg.predict(X_test)
plt.scatter(X_test, y_test)
plt.plot(X_test, y_prediksi, c='r')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Plot Hours vs Scores')
```
